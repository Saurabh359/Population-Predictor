## 3. Pie-chart

```python
def pie_charts(feature,countries,ex,color):
    tea=df.loc[countries,feature]
    temp=pd.DataFrame({'Population':tea})
    total=temp.loc['World','Population']
    sum1=temp.sum(axis=0)
    var=sum1.loc['Population']
    var=var-total
    total1=total-var
    temp=temp.replace(to_replace=total,value=total1)
    total=temp.loc['World','Population']
    temp.rename(index={'World':'Others'},inplace=True)
    plt.pie(temp,labels=temp.index,radius=2,colors=color,startangle=-50,explode=ex,shadow=True,autopct='%0.1f%%')
    plt.show()
```

```python
send1=['China','United States','India','World','Indonesia']
ex=[0,0,0.2,0,0]
color=['blue','green','red','yellow','purple']
pie_charts('2018',send1,ex,color)
```

### 1. Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
lr=LinearRegression()
```

```python
years1=list(map(str,range(1960,2011)))
```

```python
original=[]
yer1=list(map(str,range(2011,2019)))
u=df.loc['India',yer1]
for d in u:
    original.append(d)
```

```python
original
```

```python
def predict(name,yr):
    data = df.loc[name,years1] 
    out=pd.DataFrame(data)
    out.reset_index(inplace=True)
    out.rename(columns={'index':'Years',name:'Population'},inplace=True)
    out['Years']=out['Years'].astype('int')
    out['Population']=out['Population'].astype('int')
    x=out[['Years']]
    y=out['Population']
    lr.fit(x,y)
    gh=lr.predict([[yr]])
    #a=str(gh[0])
    #we=a.split('e')
    return gh[0]
```

```python
y_predict=[]
for j in range(2011,2019):
    y_pr=predict('India',j)
    y_predict.append(y_pr)
```

```python
y_predict
```

```python
linear=mean_squared_error(original,y_predict)
print(linear)
```

### 2. Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
data = df.loc['India',years] 
aout=pd.DataFrame(data)
aout.reset_index(inplace=True)
aout.rename(columns={'index':'Years','India':'Population'},inplace=True)
aout['Years']=aout['Years'].astype('int')
aout['Population']=aout['Population'].astype('int')
x=aout['Years']
y=aout['Population']
```

```python
# degree=2
f=np.polyfit(x,y,2)
p=np.poly1d(f)
print(p)

# degree=9
f2=np.polyfit(x,y,9)
p2=np.poly1d(f2)
print(p2)
```

```python
def PlotPolly(model, independent_variable, dependent_variabble, Name):
    x_new = np.linspace(1960,2018, 100)
    y_new = model(x_new)

    plt.plot(independent_variable, dependent_variabble, '.', x_new, y_new, '-')
    plt.title('Polynomial Fit with Matplotlib for Population ~ years')
    ax = plt.gca()
    ax.set_facecolor((0.898, 0.898, 0.898))
    fig = plt.gcf()
    plt.xlabel(Name)
    plt.ylabel('Population')

    plt.show()
    plt.close()
```

```python
PlotPolly(p, aout['Years'], aout['Population'], 'years')  
PlotPolly(p2, aout['Years'], aout['Population'], 'years') 
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.pipeline import Pipeline
```

```python
def polyd(country,ye,i):
    pr=PolynomialFeatures(degree=i)
    data = df.loc[country,years1] 
    about=pd.DataFrame(data)
    about.reset_index(inplace=True)
    about.rename(columns={'index':'Years',country:'Population'},inplace=True)
    about['Years']=about['Years'].astype('int')
    about['Population']=about['Population'].astype('int')
    x=about['Years']
    y=about['Population']
    
    x_train=pr.fit_transform(about[['Years']])
    lr.fit(x_train,y)
    
    x_test=pr.fit_transform([[ye]])
    gh=lr.predict(x_test)
    #a=str(gh[0])
    #we=a.split('e')
    return gh[0]
```

```python
y_pre=[]
for j in range(2011,2019):
    y_pr=polyd('India',j,23)
    y_pre.append(y_pr)
y_pre
```

```python
polynom=mean_squared_error(original,y_pre)
print(polynom)
```

```python
if polynom>linear:
    print("Linear is good")
else :
    print("Polynomial is good")
```

### Final Model Creation

```python
def poly_predict(country,ye,i):
    pr=PolynomialFeatures(degree=i)
    data = df.loc[country,years] 
    about=pd.DataFrame(data)
    about.reset_index(inplace=True)
    about.rename(columns={'index':'Years',country:'Population'},inplace=True)
    about['Years']=about['Years'].astype('int')
    about['Population']=about['Population'].astype('int')
    x=about['Years']
    y=about['Population']
    
    x_train=pr.fit_transform(about[['Years']])
    lr.fit(x_train,y)
    
    x_test=pr.fit_transform([[ye]])
    gh=lr.predict(x_test)
    return gh[0]
```

```python
def checking(country):
    original_1=[]
    u=df.loc[country,yer1]
    for d in u:
        original_1.append(d)
    prev=0
    current=0
    for i in range(2,50):
        y_pred=[]
        for j in range(2011,2019):
            y_pr=polyd(country,j,i)
            y_pred.append(y_pr)
        polynom=mean_squared_error(original_1,y_pred)
        current=polynom
        if (current>prev) & (i!=2):
            return i
        else :
            prev=current           
```

```python
def predict_population(country,year):  
    best_fit=checking(country)
    print(best_fit)
    return poly_predict(country,year,best_fit)
```

```python
input1='India'
year_input=2019
result=predict_population(input1,year_input)
print("Population of "+ input1 +" in year "+ str(year_input) +" is : "+  str(round(result, 2)) +" approx")
Line_charts(input1)
```
